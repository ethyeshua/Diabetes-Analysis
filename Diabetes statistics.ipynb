```python
# Central Tendency Calculations
print('MEAN: \n',data.mean())
print('\r \n')

print('MEDIAN: \n',data.median())
print('\r \n')

print('MODE: \n',data.mode())
print('\r \n')

print('Q1-Quartile 1: \n',data.quantile(q=(0.25,0.50,0.75)))
```

```python
mean=data['Glucose'].mean();
median=data['Glucose'].median();
mode=data['Glucose'].mode();


fig, ax = plt.subplots(figsize=(10,6));
# - - - - - - - - -setting size of the canvas

sns.distplot(data['Glucose']);
plt.title('Distribution plot of Glucose variable', color ='blue');

plt.axvline(mean, color='green', label='Mean');
plt.axvline(median, color='blue', label='Median');

#
plt.axvline(mode[0], color='red', label='Mode1')
#
plt.axvline(mode[1], color='red', label='Mode2')

plt.legend();
```

```python
# # Pairplot using s
corr = data.corr();print(corr)

# sns.heatmap(corr, annot = True);

fig, ax = plt.subplots()
fig.set_size_inches(10, 10)

plt.title('Heat Map of Correlation \n')

sns.heatmap(corr,annot=True,cmap='YlGnBu',
vmin=1,vmax=1,linewidths=.5, center=0);
```
